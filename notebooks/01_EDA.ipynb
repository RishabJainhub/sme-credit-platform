# 📊 SME Credit Risk — Exploratory Data Analysis (Jan 2026)
### *Analytical Methodology for Portfolio Risk Stratification*

This notebook explores the high-fidelity SME dataset calibrated against real-world **RBI Sectoral Deployment** and **MCA Master Data** proxies. We will analyze the geographic distribution, sectoral risk skew, and the relationship between operational factors and credit default probability.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-muted')
sns.set_context('notebook')

# Load the calibrated dataset
df = pd.read_csv('../data/sme_clean_real.csv')
df.head()
```

## 🗺️ 1. Geographic Footprint
We expanded the dataset to cover **36 States and UTs**, reflecting the diverse industrial landscape of India. Let's look at the concentration of records.

```
plt.figure(figsize=(12, 6))
df['State'].value_counts().head(15).plot(kind='bar', color='#6366F1')
plt.title('Top 15 States by SME Concentration')
plt.ylabel('Number of Records')
plt.xticks(rotation=45)
plt.show()
```

## 📊 2. Sectoral Distribution (Realistic Skew)
Unlike uniform synthetic datasets, our generation engine introduces a realistic skew where **Retail** and **Manufacturing** comprise the majority of the MSME volume.

```
plt.figure(figsize=(8, 8))
df['Sector'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['#6366F1', '#EC4899', '#A855F7', '#F59E0B', '#10B981', '#3B82F6'])
plt.title('MSME Sectoral Composition')
plt.ylabel('')
plt.show()
```

## ⚖️ 3. Risk Stratification
How does **Credit Score** relate to our engineered **Default Risk** proxy? We expect a strong inverse correlation.

```
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df, x='credit_score', hue='default_risk', fill=True, palette={0: '#10B981', 1: '#EF4444'})
plt.title('Credit Score Distribution by Risk Type (0=Stable, 1=High Risk)')
plt.show()
```

- **Sectoral skew** aligns with Jan 2026 RBI deployment trends.
